```python
from pyzomato import Pyzomato
import json
import numpy as np
import pandas as pd
import requests

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
p = Pyzomato('<ZOMATO API KEY>')

def get_restaurant_list(lat, lon, json_name):
    restaurantList = []

    for a in range(0,100,20):
        restaurants = p.search(lat=lat,lon=lon,
                               count=20,
                               sort='rating',
                               order='desc',
                               category=10,
                               start=a)
        restaurantList.append(restaurants)

    # Write list of dictionaries to a json file
    with open(json_name, 'w') as fp:
        return json.dump(restaurantList, fp)
```

```python
# sydney
sydney_data = get_restaurant_list(-33.865, 151.2094, 'Top100SydneyDinners.json')
# melbourne
melbourne_data = get_restaurant_list(-37.814, 144.96332, 'Top100MelbourneDinners.json')
```

```python
def load_json(filename):
    with open(filename) as json_file:
        allRestaurants = json.load(json_file)
        return allRestaurants
    
melb_restaurants=load_json('Top100MelbourneDinners.json');
sydney_restaurants=load_json('Top100SydneyDinners.json');

columns = ["name", "locality", "city", "cuisines", "average_cost_for_two", 'aggregate_rating', 'rating_text', 'votes']
```

```python
def restaurant_tables(city):
    restaurantTable = pd.DataFrame(columns=columns)

    for restaurantSet in city:
        restaurants = restaurantSet['restaurants']
        for restaurant in restaurants:
            
            singleRow = [restaurant['restaurant']['name'],
                   restaurant['restaurant']['location']['locality'],
                   restaurant['restaurant']['location']['city'],
                   restaurant['restaurant']['cuisines'],
                   restaurant['restaurant']['average_cost_for_two'],
                   restaurant['restaurant']['user_rating']['aggregate_rating'],
                   restaurant['restaurant']['user_rating']['rating_text'],
                   restaurant['restaurant']['user_rating']['votes']]
            newRow = pd.Series(singleRow, index=columns)

            restaurantTable = restaurantTable.append(newRow, ignore_index=True)
    return restaurantTable
 
melbourne_rest = restaurant_tables(melb_restaurants)
sydney_rest = restaurant_tables(sydney_restaurants)
```

```python
# Compute the mean for 'average_cost_for_two'
def get_costs(restaurants, city):

    costs = restaurants['average_cost_for_two']
    avgCosts = costs.mean()
    print("Average Cost for Two in " + city + " = $", avgCosts)

    # Compute percentage of values below the mean.
    belowCostMean = costs < costs.mean()
    numBelowCostMean = belowCostMean.sum()
    print("Percentage of restaurants in " + city + " below the mean is = ",numBelowCostMean*100/costs.count(), "%")
    print("-------------------------------------")
    return costs

sydney_costs = get_costs(sydney_rest, "Sydney") 
melbourne_costs = get_costs(melbourne_rest, "Melbourne") 
```

```python
# Plot the distribution
sns.set(style="ticks")
x = sns.distplot(list(melbourne_costs.values), kde=False, bins=25, color="blue")
y = sns.distplot(list(sydney_costs.values), kde=False, bins=25, color="red")

x.xaxis.set_minor_locator(plt.MultipleLocator(5))
x.xaxis.set_major_locator(plt.MultipleLocator(50))
x.yaxis.set_major_locator(plt.MultipleLocator(5))

plt.xlabel("Average cost for two ($)")
plt.ylabel("Number of Places")
plt.title("Melbourne and Sydney Top 100 Places for Dinner - V1")
plt.legend(['Melbourne - average cost for two', 'Sydney - average cost for two'])

plt.show()
```

```python
# Calculate the median and percentage of values that lie below the value.
def calculate_median(restaurants, city):
    costs = restaurants['average_cost_for_two']

    medianCost = costs.median()
    belowMedian = costs < medianCost
    numBelowMedian = belowMedian.sum()
    print("The median cost in "+ city+" is = $", medianCost)
    print("The percentage of values below the median in "+ city+ " is ", numBelowMedian*100/costs.count(), "%")
    print("---------------------------------------------------")
calculate_median(sydney_rest, "Sydney") 
calculate_median(melbourne_rest, "Melbourne") 
```

```python
# Investigate outliers
# Set the type of the average_cost_for_two column to int so we can run nsmallest() on it.

def get_new_resto_table(city):
    new_city = city.astype({'average_cost_for_two': int})
    return new_city
print("Current types: ", melbourne_rest.shape)
print("Current types: ", sydney_rest.shape)

melbourne_resto = get_new_resto_table(melbourne_rest)
sydney_resto = get_new_resto_table(sydney_rest)

print("Updated type: ", melbourne_resto.dtypes)
print("Updated type: ", sydney_resto.dtypes)
```

```python
def get_cheapest_restaurants(restaurants):
    
    cheapestRestaurants = restaurants.nsmallest(10,'average_cost_for_two')
    return cheapestRestaurants
```

```python
cheap_resto_melb = get_cheapest_restaurants(melbourne_resto)
cheap_resto_melb
```

```python
cheap_resto_sydney = get_cheapest_restaurants(sydney_resto)
cheap_resto_sydney
```

```python
# eliminates cafees and cheap restaurants

# Create a list of cuisines we do not want, that will be used to filter out the cafes from the dataset
def filter_cuisines(restaurants):
    discardCuisines = ['Ice Cream', 'Coffee and Tea', 'Cafe Food']
    cuisines = restaurants['cuisines']

    discardRestaurants = cuisines.str.contains('|'.join(discardCuisines))
    filteredRestaurants = discardRestaurants == False
    restaurants[discardRestaurants]
    restaurants = restaurants[filteredRestaurants]
    print("The size of our updated dataset is ", restaurants.shape)
    return restaurants
```

```python
melbourne_resto_filtered = filter_cuisines(melbourne_resto)
sydney_resto_filtered = filter_cuisines(sydney_resto)
```

```python
# We compute some descriptive statistics again.

def compute_updated(restaurants, city):
    costs = restaurants.average_cost_for_two
    averageCost = costs.mean()
    medianCost = costs.median()
    belowMedian = costs < medianCost
    numBelowMedian = belowMedian.sum()
    print("The average cost in " + city + " is = $", averageCost)
    print("The median cost is = $", medianCost)
    print("The percentage of values below the median is ", numBelowMedian*100/costs.count(), "%")
    print("----------------------------------------------------")
    return costs

melb_costs = compute_updated(melbourne_resto_filtered, "Melbourne")
sydney_costs = compute_updated(sydney_resto_filtered, "Sydney")
```

```python
# Plot the distribution
sns.set(style="ticks")
x = sns.distplot(list(melb_costs.values), kde=False, bins=25, color="blue")
y = sns.distplot(list(sydney_costs.values), kde=False, bins=25, color="green")

x.xaxis.set_minor_locator(plt.MultipleLocator(5))
x.xaxis.set_major_locator(plt.MultipleLocator(50))
x.yaxis.set_major_locator(plt.MultipleLocator(5))

plt.xlabel("Average cost for two ($)")
plt.ylabel("Number of Places")
plt.title("Melbourne and Sydney Top 100 Places for Dinner - V2")
plt.legend(['Melbourne - average cost for two', 'Sydney - average cost for two'])

plt.show()
```
